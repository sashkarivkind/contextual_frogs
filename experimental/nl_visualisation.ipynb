```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import numpy as np
import sys

# Specify the directory you want to add
custom_path = './..'

# Add the directory to sys.path
if custom_path not in sys.path:
    sys.path.append(custom_path)


from models import MLP, SymReLU, OneOverSqr
def plot_nonlinearity(nonlinear_fn, x_range=(-4, 4), title="Nonlinearity Visualization"):
    """
    Plots a PyTorch nonlinearity function over a given x-range.

    Args:
        nonlinear_fn: PyTorch nonlinearity, e.g., nn.Sigmoid(), nn.ReLU(), etc.
        x_range (tuple): The range of x values to evaluate the function.
        title (str): Title of the plot.
    """
    # Generate input data
    x = torch.linspace(x_range[0], x_range[1], 1000)
    
    # Pass data through the nonlinearity
    y = nonlinear_fn(x)

    # Plot
#     plt.figure(figsize=(8, 6))
    plt.plot(x.numpy(), y.numpy(), label=nonlinear_fn.__class__.__name__)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title(title)
    plt.axhline(0, color='gray', linestyle='--', linewidth=0.7)
    plt.axvline(0, color='gray', linestyle='--', linewidth=0.7)
    plt.legend()
    plt.grid(True)
#     plt.show()

# Example usage
plot_nonlinearity( nn.Tanh(), x_range=(-2, 2))
plot_nonlinearity(OneOverSqr(), x_range=(-2, 2))
plot_nonlinearity(SymReLU(), x_range=(-2, 2))
```
