```python
def local_eval_paradigms(model, playlist_file, args, include_prefs=None):
    with open(playlist_file, 'rb') as f:
        playlist = pickle.load(f)

    model.eval()
    with torch.no_grad():
        outputs = {}
        for paradigm_name,paradigm in playlist.items():
            if include_prefs is None or np.any([pref in paradigm_name  for pref in include_prefs]):
                print(f'evaluating paradigm {paradigm_name} with {len(paradigm)} steps')
                ys = torch.tensor(paradigm, device=next(model.parameters()).device)
                noises = torch.randn((args.bs, len(ys)), device=next(model.parameters()).device) * model.sigma_x  # [bs, t]
                noises = [noises[:, t_idx] for t_idx in range(noises.shape[1])]
                model_setting = args.model
                outputs_ = model.f(args.n,
                                    noises,
                                    ys,  
                                    model_setting,
                                    qs=None,
                                    )
                outputs[paradigm_name] = np.array([z.cpu().numpy().reshape(-1) for z in outputs_]) 
    return outputs
```

```python
# def forano_trial(times = [160,752,112,320], pert=[0,1,-1,np.nan]):
def forano_trial(times = [160,960,200,320], pert=[0,1,-1,np.nan]):
    ymask  = np.concatenate([p*np.ones(t) for p,t in zip(pert, times)]) 
    qs = []
    for t in times:
        v = np.random.random(size=t)
        these_qs = v > np.median(v)
        these_qs = these_qs*2 -1
        qs.append(these_qs)
    qs = np.concatenate(qs)
    ys = ymask * qs
    return ys, qs
    
```
