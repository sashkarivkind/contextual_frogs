```python
P0=0
Pplus=1
Pminus=-1
Pchannel=np.nan

T = 1500
Tants = [200,500,800]
playlist = {**{f'antero{i}': [(P0,1000),(Pplus,Tant),(Pminus,1000)] for i,Tant in enumerate(Tants)},
'frogs': [(P0,T),((P0,Pplus),T),(P0,T)],
'anti_frogs': [(P0,T),(Pplus,T),(P0,T)],
'savings': 3 * [(P0, 500), (Pplus, 500)]}
# 'sprecovery': [(P0, 192),
#                (Pplus, 384),
#                (Pminus, 20),
#                (Pchannel, 400)],
# 'spOVRrecovery': [(P0, 192),
#                (Pplus, 1000),
#                (Pminus, 20),
#                (Pchannel, 400)],
# 'evrecovery': [(P0, 192),
#                (Pplus, 384),
#                (Pminus, 20),
#                (Pplus, 20),
#                (Pchannel, 400)],
# 'lng_recovery': [(P0, 192),
#                (Pplus, 384),
#                (Pminus, 500),
#                (Pchannel, 400)]
#     }
```

```python
baseline_settings = {'criterion':'MSE', 'k':[0.5,0,1], 'learning_rate':1e-3, 'sigma_noi':0.0, 'tau_u':10}
ff_settings = {'criterion':'MSE', 'k':[0,1,0], 'learning_rate':1e-3, 'sigma_noi':0.0, 'tau_u':10}

model_construct_args =  dict(n_inputs = 3,
                  n_hidden = 4*512,
                  n_outs = 1,
                  en_bias = False,
                 b_high=3, ones_first_layer=True,
                  nl = lambda : SymReLU(threshold=0.4))

wide_model_construct_args =  dict(n_inputs = 3,
                  n_hidden = 4*512,
                  n_outs = 1,
                  en_bias = False,
                 b_high=3, ones_first_layer=True,
                  nl = lambda : SymReLU(threshold=2.5))

test_vec= np.stack([np.linspace(-10,10,200), 
                    np.zeros(200),
                    np.zeros(200)]).T
```

```python
baseline_runner = Runner(model_construct_args=model_construct_args,
                test_vec=test_vec,
                **baseline_settings)
wide_runner = Runner(model_construct_args=wide_model_construct_args,
                test_vec=test_vec,
                **{**baseline_settings, 'learning_rate':0.5e-7 })
```

```python
ff_runner = Runner(model_construct_args=model_construct_args,
                test_vec=test_vec,
                **ff_settings)
```

```python
ffwide_runner = Runner(model_construct_args=wide_model_construct_args,
                test_vec=test_vec,
                **{**ff_settings, 'learning_rate':0.5e-7 })
```

```python
results = baseline_runner.run_multiple(playlist)
```

```python
ff_results = ff_runner.run_multiple(playlist)
```

```python
wide_results = wide_runner.run_multiple(playlist)
```

```python
ffwide_results = ffwide_runner.run_multiple(playlist)
```

```python
def do_plots(results,playlist,figsize=(24, 6)):
    plt.figure(figsize=figsize)
    plt.subplot(1,3,1)
    plot_segments(results['savings'].u_lp,
              t_start=playlist['savings'][0][1],
              n_segments=3,
              t_increment=playlist['savings'][0][1]+playlist['savings'][1][1],
              t_segment=playlist['savings'][1][1],             
             )
    plt.subplot(1,3,2)
    plot_by_key(results, ['frogs','anti_frogs'],visu_offsets = [0, 0.0])
    plt.xlabel('epochs')
    plt.ylabel('u(epoch) [au]')
    plt.subplot(1,3,3)
    plot_by_key(results,
                [f'antero{i}'for i,Tant in enumerate(Tants)],
                colors=['tab:red','tab:blue','tab:orange','tab:green'],
                align_end=True,
                visu_offsets=[0]*10,
               )
```

```python
do_plots(results,playlist)
plt.subplot(1,3,1)
plt.title(f'savings')
plt.subplot(1,3,2)
plt.title(f'abrupt/gradual')
plt.subplot(1,3,3)
plt.title(f'interference')
plt.grid(True)
plt.show()


do_plots(ff_results,playlist)
plt.subplot(1,3,1)
plt.title(f'savings')
plt.subplot(1,3,2)
plt.title(f'abrupt/gradual')
plt.subplot(1,3,3)
plt.title(f'interference')
plt.grid(True)
plt.show()

do_plots(wide_results,playlist)
plt.subplot(1,3,1)
plt.title(f'savings')
plt.subplot(1,3,2)
plt.title(f'abrupt/gradual')
plt.subplot(1,3,3)
plt.title(f'interference')
plt.grid(True)
plt.show()

do_plots(ffwide_results,playlist)
plt.subplot(1,3,1)
plt.title(f'savings')
plt.subplot(1,3,2)
plt.title(f'abrupt/gradual')
plt.subplot(1,3,3)
plt.title(f'interference')
plt.grid(True)
plt.show()

```
